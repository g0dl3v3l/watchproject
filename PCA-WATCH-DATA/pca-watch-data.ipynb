```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
columns = ['ax', 'ay', 'az', 'gx', 'gy', 'gz','ma', 'mg']
```

```python
# Hi motion performed by person A
dataset1 = pd.read_csv('handwave_righthand_dheeraj_trial_1.csv')
dataset1.drop(columns=['timePeriod', 'label', 'TIME', 'DATE'], inplace=True)
dataset1 = pd.DataFrame(preprocessing.normalize(dataset1, axis=0), columns=columns)
dataset1.head()
```

```python
# Hi motion performed by person B
dataset2 = pd.read_csv('handwave_righthand_jshwant_trial_2.csv')
dataset2.drop(columns=['timePeriod', 'label', 'TIME', 'DATE'], inplace=True)
dataset2 = pd.DataFrame(preprocessing.normalize(dataset2, axis=0), columns=columns)
dataset2.head()
```

```python
# left right swipe performed by person A
dataset3 = pd.read_csv('left-right_swip_righthand_dheeraj_trial_1.csv')
dataset3.drop(columns=['timePeriod', 'label', 'TIME', 'DATE'], inplace=True)
dataset3 = pd.DataFrame(preprocessing.normalize(dataset1, axis=0), columns=columns)
dataset3.head()
```

```python
# right left swipe performed by person A
dataset4 = pd.read_csv('right-left_swip_righthand_dheeraj_trial_1.csv')
dataset4.drop(columns=['timePeriod', 'label', 'TIME', 'DATE'], inplace=True)
dataset3 = pd.DataFrame(preprocessing.normalize(dataset1, axis=0), columns=columns)
dataset4.head()
```

```python
datasets = [dataset1, dataset2, dataset3, dataset4]
for dataset in datasets:
    print(f"shape: {dataset.shape}")
```

```python
# total time performed the motion is 60 sec
# the number of samples per motion is 1220
# by taking 1220 points on x-axis we are plotting each of the col data on y-axis
```

```python
# dataset 1 - accelerometer
data_points = 1220
plt.figure(figsize=(25, 25))

plt.subplot(4, 1, 1)
plt.title("accelerometer x-cordinate vs each datapoint", fontdict={'fontsize': 18})
plt.plot(range(data_points), dataset1['ax'],  color="green")

plt.subplot(4, 1, 2)
plt.title("accelerometer y-cordinate vs each datapoint", fontdict={'fontsize': 18})
plt.plot(range(data_points), dataset1['ay'],  color="green")

plt.subplot(4, 1, 3)
plt.title("accelerometer z-cordinate vs each datapoint", fontdict={'fontsize': 18})
plt.plot(range(data_points), dataset1['az'],  color="green")

plt.subplot(4, 1, 4)
plt.title("accelerometer norm vs each datapoint", fontdict={'fontsize': 18})
plt.plot(range(data_points), dataset1['ma'],  color="green")
```
